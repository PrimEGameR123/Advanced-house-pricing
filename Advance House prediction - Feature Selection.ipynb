```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
pd.pandas.set_option('display.max_columns',None)
```

```python
ds=pd.read_csv('x_train.csv')
```

```python
ds.head()
```

```python
y_train=ds[['SalePrice']]
```

```python
y_train
```

```python
x_train=ds.drop(['Id','SalePrice'],axis=1)
```

```python
### applying feature selection
# First, I specify the Lasso Regression model, and 
## select a suitable alpha (equivalent of penalty)
### The bigger the aphla, lesser the features that will be selected

# Then I use the SelectFromModel object from sklearn, which will select
## the features which coefficient are non-zero

feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(x_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
selected_fe=x_train.columns[(feature_sel_model.get_support())]
selected_fe
```

```python
print('Total features : {}'.format(x_train.shape[1]))
print('Selected features : {}'.format(len(selected_fe)))
print('Features shrink to zero : {}'.format(np.sum(feature_sel_model.estimator_.coef_==0)))
```
